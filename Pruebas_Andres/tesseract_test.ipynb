# First tests with tesseract

This test reads the text of the images that have not been answered and in the time interval for answering the questions. That's because the library cannot understand text with a colores background 

```python
from PIL import Image
import pytesseract
```

```python
imagen = Image.open("imagenes/ima6.png")
imagen
```

```python
text = pytesseract.image_to_string(imagen, lang = "spa")
text
```
